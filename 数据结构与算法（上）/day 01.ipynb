2. 托普利茨矩阵问题

如果一个矩阵的每一方向由左上到右下的对角线上具有相同元素，那么这个矩阵是托普利茨矩阵。

给定一个M x N的矩阵，当且仅当它是托普利茨矩阵时返回True。

```python
def isToeplitzMatrix(matrix):
    for i in range(len(matrix)-1):
        if not matrix[i][:-1] == matrix[i+1][1:]:
            return False
    return True
```

```python
matrix = [
  [1,2,3,4],
  [5,1,2,3],
  [9,5,1,2]
]
isToeplitzMatrix(matrix)
```

3. 三数之和
给定一个包含 n 个整数的数组nums，判断nums中是否存在三个元素a，b，c，使得a + b + c = 0？找出所有满足条件且不重复的三元组。

```python
def threesum(numbers):
    res = []
    numbers.sort()
    print(numbers)
    i = 0
    j = len(numbers)
    if j < 3:
        return False
    for i in range(j):
        left = i + 1
        right = j -1
        while(left<right):
            temp_res = numbers[i] +numbers[left] + numbers[right]
            if temp_res == 0 :
                res.append([numbers[i],numbers[left],numbers[right]])
                left += 1
                right -=1
            elif temp_res > 0:
                left += 1
            else:
                right -=1     
    return res
```

```python
nums = [-1, 0, 1, 2, -1, -4]
```

```python
threesum(nums)
```
