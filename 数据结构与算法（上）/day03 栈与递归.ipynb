```python
#用数组实现一个顺序栈
class arrayStack:
    def __init__(self):
        self.stack = []
        
    def isempty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False
    
    def append(self,x):
        self.stack.append(x)
        
    def pop(self):
        self.stack.pop()
    
    def peak(self):
        if self.is_empty:
            return None
        return self.stack[-1]
    
```

```python
#用链表实现一个链栈
class linkNode:
    def __init__(self,data,next_ = None):
        self.head = data
        self.next = next_
class linkArray:
    def __init__(self):
        self._top = None
    
    def push(self,data):
        self._top = linkNode(data,self._top)
    def pop(self):
        if self._top == None:
            return None
        self._top = self._top.next
        return self._top.data
    def top(self):
        if self._top == None:
            return None
        else:
            return self._top.data
        
```
