```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        val1 = l1.val
        val2 = l2.val
        pre1 = l1.next
        pre2 = l2.next
        while pre1 and pre2:
            if val1 <= val2:
                pre1 = pre1.next
            else:
                temp = pre1.next
                pre1.next = pre2
                pre1 = temp.next
        while pre1 == None and pre2:
            pre2 = pre2.next
        while pre2 == None and pre1:
            pre1 = pre1.next
        
```

2. 删除链表的倒数第N个节点

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

def length(head):
    p,n = head,0
    while p is not None:
        n += 1
        p = p.next
    return n  
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        k = 0
        pre,temp = head, head
        leng = length(head)
        if n <0 or n > length(head):
            return False
        while(k<leng-n and pre.next):
            pre = pre.next
            k +=1
        while(pre.next):
            pre = pre.next
            temp = temp.next
        temp.next = temp.next.next
        return head
```

3. 旋转链表
给定一个链表，旋转链表，将链表每个节点向右移动k个位置，其中k是非负数。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

def length(head):
    p,n = head,0
    while p is not None:
        n += 1
        p = p.next
    return n  

class Solution:
    def rotateRight(self, head: ListNode, k: int) -> ListNode:
        leng = length(head)
        if leng == 0:
            return head
        n = k%leng
        if n == 0 or leng == 1:
            return head
        slow,fast = head,head
        while(fast.next and n > 0):
            n -= 1
            fast = fast.next
        while(fast.next):
            fast = fast.next
            slow = slow.next
        new_head = slow.next
        fast.next = head
        slow.next = None
        return new_head
        
```
